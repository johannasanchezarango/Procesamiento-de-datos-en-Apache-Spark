```python
#Importamos las librerias 
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
#Cargamos el dataset y mostramos las 5 primeras filas
data = pd.read_csv("C:/Users/nana_/OneDrive/Desktop/UNAD/2025/Dataset Arboles de Decisión/winequality-red.csv")
data.head()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=data[['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar', 'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density', 'pH','sulphates', 'alcohol']])
plt.xticks(rotation=45)
plt.title('Boxplot de calidad del vino')
plt.ylabel('Valores')
plt.show()
```

```python
predictors_col = ["fixed acidity", "volatile acidity", "citric acid", "residual sugar", "chlorides", "free sulfur dioxide", "total sulfur dioxide", "density", "pH", "sulphates", "alcohol"]
target_col = ['quality']
```

```python
predictors = data[predictors_col]
target = data[target_col]
```

```python
predictors
```

```python
target
```

```python
corr = predictors.corr()
corr
```

```python
sns.heatmap(corr)
plt.title('Matriz de Correlación')
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(predictors, target, test_size=0.2, random_state=13)
```

```python
tree = DecisionTreeClassifier(max_depth=3, random_state=42)
```

```python
arbol = tree.fit(X_train, y_train)
```

```python
plot_tree(arbol)
```

```python
predicciones = arbol.predict(X_test)
```

```python
predicciones
```

```python
# Examinamos la matriz de confusión
pd.crosstab(np.array([y[0] for y in y_test.values.tolist()]), predicciones, rownames=['Actual'], colnames=['Predicciones'])
```

```python
accuracy = accuracy_score(y_test,predicciones)
accuracy
```

```python
precision = precision_score(y_test, predicciones, zero_division=0, average='weighted')
precision
```

```python
recall = recall_score(y_test, predicciones, average='weighted')
recall
```

```python
0.9666666666666667
f1 = f1_score(y_test, predicciones, average='weighted')
f1
```

## Interpretar, analizar y documentar los resultados obtenidos.

## 1. Métricas de rendimiento
Exactitud (Accuracy): El modelo alcanzó una precisión global del 57.8%, lo que indica que aproximadamente 6 de cada 10 muestras fueron clasificadas correctamente.

Precisión (Precision): Con un valor ponderado de 54.2%, el modelo tiene una moderada capacidad para evitar falsos positivos, aunque podría ser mejor en términos de discriminación entre clases.

Exhaustividad (Recall): El modelo logró un 57.8%, lo que refleja su habilidad para identificar correctamente las clases reales, pero deja espacio para mejorar la identificación de muestras de ciertas categorías.

F1-Score: El equilibrio entre precisión y recall se refleja en un F1-Score ponderado de 54.2%, lo que sugiere una clasificación efectiva en clases predominantes, pero un bajo desempeño en clases menos representadas.

## 2. Matriz de Confusión
La matriz de confusión proporciona una visión más detallada del desempeño del modelo:

Clase 5: Mayor cantidad de verdaderos positivos (125), pero también algunos clasificados incorrectamente como clase 6.

Clase 6: 68 verdaderos positivos, pero con una alta cantidad de instancias clasificadas como clase 5.

Clases 3 y 8: Estas clases tienen muy pocas observaciones correctamente clasificadas, evidenciando dificultades en la predicción de clases minoritarias.

## 3. Matriz de correlación:
El análisis sugiere que las variables como alcohol y density tienen una influencia significativa en la calidad del vino. Sin embargo, algunas correlaciones fuertes pueden llevar a redundancia en los datos.

## 4. Árbol de decisión:
